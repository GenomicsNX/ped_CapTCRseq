### Start with estimate manifest df from IPD study and subset to cbttc

```r
load("/Users/anabbi/OneDrive - UHN/Documents/ImmPedCan/Finaldata/ESTIMATE/total_estimate_manifest.RData")
```

```r
cbttc <- total_estimate_manifest[ total_estimate_manifest$group == "CBTTC",]
```

### find case_ids that have recurrence events

```r
case_idrecur <- cbttc$case_id[ cbttc$sample_type == "Recurrence"]
```

```r
sampletypes_recur <- cbttc[ cbttc$case_id %in% case_idrecur,]
```

### remove progressive samples

```r
cbttc_recurr <- sampletypes_recur[, c("sample_type", "case_id", 
                             "sample_id", "disease_type")]

cbttc_recurr_noprog <- cbttc_recurr[cbttc_recurr$sample_type != "Progressive",]

cbttc_recurr_noprog[order(cbttc_recurr_noprog$case_id),]
```

### Remove unique IDs to get only initial CNS tumor/recurrence pairs

```r
cbttc_recurr_noprog_dupids <- cbttc_recurr_noprog$case_id[duplicated(cbttc_recurr_noprog$case_id)]
```

```r
initalrecurs <- cbttc_recurr_noprog[cbttc_recurr_noprog$case_id %in% cbttc_recurr_noprog_dupids,]
```

```r
initalrecurs[order(initalrecurs$case_id),]
```

### C271092 has two recurrence events and no initial tumor. Remove it!

```r
initalrecurs_recrm <- initalrecurs[initalrecurs$case_id != "C271092",]
```

```r
initalrecurs_recrm
```

### Load original manifest file by JM to include age at collection

```r
clinical <- read.csv("/Users/anabbi/OneDrive - UHN/Documents/ImmPedCan/Finaldata/Manifests/CBTTC_clinical_dedup.csv",
                    sep = ",", header = T)
```

```r
cbttc_ageatcollection <- clinical[, c("CBTTC.Subject.ID", "CBTTC.Event.ID",
                                      "Kids.Subject.ID", "Age.at.Collection")]

colnames(cbttc_ageatcollection) <- c("case_id", "sample_id",
                                    "patient_id", "Age_at_collection")
```

```r
initalrecurs_recrm_age <- merge(initalrecurs_recrm, cbttc_ageatcollection,
                               by = "sample_id")
```

### Order based on disease type for plot

```r
initalrecurs_recrm_age <- initalrecurs_recrm_age[order(initalrecurs_recrm_age$case_id.x),]
initalrecurs_recrm_age <- initalrecurs_recrm_age[order(initalrecurs_recrm_age$disease_type),]
```

```r
initalrecurs_recrm_age
```

```r
initalrecurs_recrm_age$case_id.x <- as.factor(initalrecurs_recrm_age$case_id.x)
```

```r
initalrecurs_recrm_age$case_id.x <- factor(initalrecurs_recrm_age$case_id.x,
                                          levels = c("C182778", "C300981", "C121647",
                                                     "C20172", "C253872", "C35793", "C47724",
                                                     "C42804", "C43542", "C62484", "C94956",
                                                     "C38868", "C73062", "C75276", "C77490",
                                                     "C88683", "C27429",
                                                     "C51906", "C70110", "C879327"))
```

```r
initalrecurs_recrm_age$case_id.x 
```

```r
initalrecurs_recrm_age$Age_at_collection <- as.numeric(as.character(initalrecurs_recrm_age$Age_at_collection))
```

```r
pairs_p <- ggplot(aes(x = Age_at_collection, y = case_id.x), data = initalrecurs_recrm_age) + 
geom_point(size = 3) + mytheme +
geom_line(aes(group = case_id.x), size = 1) +
scale_x_continuous(limits = c(0,9000),
                  breaks = seq(0,9000, by = 365),
                  labels = seq(0,9000/365, by = 1)) +
scale_y_discrete(labels = c("CHR", "DNET", "EP", "EP",
                            "EP", "EP", "EP", "EWS",
                            "GGM", "GGM", "HGG", "LGG",
                            "LGG", "LGG", "LGG", "LGG",
                            "MB", "MN", "MN", "RDR/SAR"))
```
