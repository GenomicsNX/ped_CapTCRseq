```r
divplot.fx <- function(divdf, filenameorder, plotname){
    
# Function to align plots (from stackoverflow) 
    align_plots1 <- function (...) {
    pl <- list(...)
    stopifnot(do.call(all, lapply(pl, inherits, "gg")))
    gl <- lapply(pl, ggplotGrob)
    bind2 <- function(x, y) gtable:::rbind_gtable(x, y, "first")
    combined <- Reduce(bind2, gl[-1], gl[[1]])
    wl <- lapply(gl, "[[", "widths")
    combined$widths <- do.call(grid::unit.pmax, wl)
    grid::grid.newpage()
    grid::grid.draw(combined)
    }

    mytheme <- theme(axis.title.y = element_text(size = 25),
                     axis.title.x = element_blank(),
                     axis.line = element_line(color = "black"),
                     axis.text = element_text(size = 22),
                     axis.text.x = element_text(angle = 45, hjust = 1)) +
    theme(panel.grid.major = element_blank(),
          panel.grid.minor = element_blank(),
          panel.background = element_rect(fill = "transparent",colour = NA),  
          legend.key = element_rect(fill = "white", colour = "white"),
          plot.margin = unit(c(0,0,0,0),"cm"))    
    

    divdf$filename <- factor(divdf$filename, levels = filenameorder)
    
#plot CPKR
    plot_CPKR <- ggplot(aes(x = filename, y = CPKR), 
                                  data = divdf) + 
    geom_point(size = 5) + 
    mytheme + theme(legend.position = "none",
                   axis.text.x = element_blank(),
                   axis.title.x = element_blank()) +
    geom_line(aes(group = case_id, color = case_id), size = 1)   
    

#plot Richness
    plot_rich <- ggplot(aes(x = filename, y = observed_Richness),          
                                data = divdf) + 
    geom_point(size = 5) + 
    mytheme + theme(legend.position = "none",
                   axis.text.x = element_blank(),
                   axis.title.x = element_blank()) +
    geom_line(aes(group = case_id, color = case_id), size = 1)   

    
#plot Shannon diversity 
    plot_shan <- ggplot(aes(x = filename, y = observed_Shannon), 
                                  data = divdf) + 
    geom_point(size = 5) + 
    mytheme + theme(legend.position = "none",
                   axis.text.x = element_blank(),
                   axis.title.x = element_blank()) +
    geom_line(aes(group = case_id, color = case_id), size = 1)   
    

#plot Simpson div
    plot_simp <- ggplot(aes(x = filename, y = observed_Simpson), 
                                  data = divdf) + 
    geom_point(size = 5) + 
    mytheme + theme(legend.position = "none",
                   axis.text.x = element_blank(),
                   axis.title.x = element_blank()) +
    geom_line(aes(group = case_id, color = case_id), size = 1)     

#plot inverse simpson    
    plot_invsimp <- ggplot(aes(x = filename, y = 1/observed_Simpson), 
                                  data = divdf) + 
    geom_point(size = 5) + 
    mytheme + theme(legend.position = "none") +
    geom_line(aes(group = case_id, color = case_id), size = 1)     
    
    
    pdf(file = paste(plotpath, plotname, sep = ""),
        width = 15, 
        height = 20,
        useDingbats = FALSE,
        onefile = FALSE)
    align_plots1(plot_CPKR + ylab("Clones per \n1000 reads"),
                 plot_rich + ylab("Richness"),
                 plot_shan + ylab("Shannon diversity"),
                 plot_simp + ylab("Simpson diversity"),
                 plot_invsimp + ylab("Inverse Simpson diversity"))
    dev.off()         
}
```
