# Repertoire overlap using immunarch

```r
immdata <- repLoad(datapath)
```

```r
TRBrep <- immdata[grep("CLONES_TRB",names(immdata))]
```

```r
TRBrep_ds <- TRBrep[grep("2000000",names(TRBrep))]

TRBrep_ds_CHP <- TRBrep_ds[grep("CHP",names(TRBrep_ds))]
```

```r
names(TRBrep_ds_CHP) <- gsub("-PBMC-DNA_2000000", "", names(TRBrep_ds_CHP))
names(TRBrep_ds_CHP) <- gsub("CLONES_TRB", "", names(TRBrep_ds_CHP))
```

```r
names(TRBrep_ds_CHP)
```

# Public repertoire

```r
pr.aa = pubRep(TRBrep_ds_CHP, "aa", .verbose = F)
```

# How many out-of-frame cdr3?

```r
mycdr3s <- pr.aa$CDR3.aa
```

```r
outofframes <- mycdr3s[grepl("~", mycdr3s)]
stopcodons <- mycdr3s[grepl("[*]", mycdr3s)]
```

```r
length(outofframes)
length(stopcodons)
```

```r
outofframes
```

# Remove outofframes and those containing stop codons

```r
pr.aa_clean <- pr.aa[!pr.aa$CDR3.aa %in% outofframes &
                     !pr.aa$CDR3.aa %in% stopcodons,]
```

```r
dim(pr.aa_clean)
```

```r
head(pr.aa_clean)
```

```r
pr.aa_clean$Samples <- NULL
```

```r
pr.aa_clean_t <- t(pr.aa_clean)
```

```r
colnames(pr.aa_clean_t) <- pr.aa_clean_t[rownames(pr.aa_clean_t) == "CDR3.aa",]
pr.aa_clean_t <- as.matrix(pr.aa_clean_t[rownames(pr.aa_clean_t) != "CDR3.aa",])
```

```r
class(pr.aa_clean_t) <- "numeric"
```

```r
pr.aa_clean_t <- as.data.frame(pr.aa_clean_t)
```

```r
pr.aa_clean_t$patient_id <- substr(rownames(pr.aa_clean_t), 1, 
                                   nchar(rownames(pr.aa_clean_t))-3)
pr.aa_clean_t$cycle <- substr(rownames(pr.aa_clean_t), 
                              nchar(rownames(pr.aa_clean_t))-1, 
                              nchar(rownames(pr.aa_clean_t)))
```

```r
col_cdr3 <- colnames(pr.aa_clean_t)[colnames(pr.aa_clean_t) != "patient_id"&
                                    colnames(pr.aa_clean_t) != "cycle"]
```

```r
pr.aa_clean_t <- pr.aa_clean_t[,c("patient_id", "cycle", col_cdr3)]
```

```r
head(pr.aa_clean_t)
```

```r
na_cols <- colSums(is.na(pr.aa_clean_t))
#na_cols[which(na_cols > 20)]
```

```r
privates <- na_cols[which(na_cols > 95)]

length(privates)
```

# private clonotypes removed

```r
privates_rm <- pr.aa_clean_t[, !colnames(pr.aa_clean_t) %in% names(privates)]
```

```r
head(privates_rm)
```

# Number of samples and patients for each common clonotype

```r
mycdr3matrix <- matrix(ncol = 2, nrow = 138,
                      dimnames = list(colnames(privates_rm)[3:ncol(privates_rm)],
                                     c("Samples", "Patients")))
```

```r
for(i in 1:nrow(mycdr3matrix)){
    mycdr3 <- rownames(mycdr3matrix)[i]
    tmp <- privates_rm[, c("patient_id", mycdr3)]
    tmp_narm <- tmp[!is.na(tmp[[mycdr3]]),]
    mycdr3matrix[i,1] <- nrow(tmp_narm)
    unqs <- unique(tmp_narm$patient_id)
    mycdr3matrix[i,2] <- length(unqs)

}
```

```r
mycdr3matrix
```

```r
mycdr3matrix <- as.data.frame(mycdr3matrix)
```

# Common clonotypes

```r
commons <- mycdr3matrix[mycdr3matrix$Patients > 1, ]
```

```r
dim(commons)
```

```r
commons
```
