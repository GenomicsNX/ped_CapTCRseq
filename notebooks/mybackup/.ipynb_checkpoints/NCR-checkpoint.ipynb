# FUNCTION heatmap of NCRs

```r
heatmap_NCR_fx <- function(datapath, NCRdffile, patient, cyclelist, plotpath){
    NCRs <- read.csv(paste(datapath, NCRdffile, sep = ""),
                 header = T, 
                 check.names = F)
    mypt <- NCRs[NCRs$Patient == patient,]                     
    mypt_mycycles <- mypt[mypt$cycle %in% cyclelist,]
    rownames(mypt_mycycles) <- mypt_mycycles$cycle                    
    NCRmat <- as.matrix(mypt_mycycles[,c("PD1%", "PDL1%", "LAG3%",
                                           "TIM3%", "CTLA4%")])   
                  
    col_fun = colorRamp2(c(0, 25, 50), 
                         c("white","#ef6548","#7f0000"))

    anno_CD3 = HeatmapAnnotation(totalCD3 = anno_barplot((mypt_mycycles$CD3)),
                                 height = unit(2,"cm"))
    heatmap_ncr = Heatmap(t(NCRmat),
                          top_annotation = anno_CD3,
                          name = "NCR",
                          column_title = paste(patient, unique(mypt_mycycles$Disease_type), sep = "_"),                             
                          col = col_fun,
                          #format rows
                          row_names_gp = gpar(fontsize = 15),                      
                          cluster_rows = FALSE,
                          show_row_dend = FALSE,                      
                          #format cols
                          cluster_columns = FALSE,
                          column_names_gp = gpar(fontsize = 15), 
                          rect_gp = gpar(col = "black"), 
                          #legend
                          show_heatmap_legend = TRUE,
                          width = unit(length(cyclelist), "cm"),
                          height = unit(5, "cm"),
                          heatmap_legend_param = list(labels_gp = gpar(fontsize = 12)))
 #   draw(heatmap_ncr)
    pdf(file = paste(plotpath, "heatmap_NCR_", patient, ".pdf", sep = ""),
                     width = 10, 
                     height = 10,
                     useDingbats = FALSE)
    print(heatmap_ncr)
    dev.off()
                         
}
```
