#### ICI

```r
mycycle_ICI <- mycycle[,c("PD1%", 'LAG3%', 'TIM3%')]
rownames(mycycle_ICI) <- mycycle$Patient
```

## Any apparent pattern of diversity changes in samples with the first two cycles?

```r
myset <- masterfile_div_trb[masterfile_div_trb$cycle %in% c("01", "02"),]
```

```r
samplefreq <- as.data.frame(table(myset$Patient))
```

```r
mysampleset <- samplefreq$Var1[ samplefreq$Freq > 1]
```

```r
masterfile_div_trb_complete <- myset[myset$Patient %in% mysampleset,]
```

```r
dim(masterfile_div_trb_complete)
```

```r
ggplot(data = masterfile_div_trb_complete, aes(x = cycle, y = log10(observed_Shannon))) +
geom_line(aes(group = Patient, color = Disease_type)) + geom_point() + mytheme
```

```r
mydf <- masterfile_div_trb_complete[duplicated(masterfile_div_trb_complete$Patient),]
```

```r
mydf$Diff_div <- NULL
```

```r
for(i in 1:nrow(mydf)){
    mypt <- mydf$Patient[i]
    temp <- masterfile_div_trb_complete[ masterfile_div_trb_complete$Patient == mypt,]
    mydf$Diff_div[i] <- log10(temp$observed_Shannon[temp$cycle == "02"]) - log10(temp$observed_Shannon[temp$cycle == "01"])
}
```

```r
summary(merged_df$FOLD)
```

```r
merged_df$Foldgroup[merged_df$FOLD < 2] <- "Fail"

merged_df$Foldgroup[merged_df$FOLD > 5] <- "Pass"
```

```r
table(merged_df$cycle)
```

```r
leukemias <- c("ALL", "AML", "SR ALL", "HR ALL", "T-ALL", "CML")
lymphomas <- c("ALCL", "BL", "HD")
solids <- c("HB", "ES", "WT", "ARMS", "ERMS", "OS", "NB")
```

```r
merged_df$Dxgroup[merged_df$Disease_type %in% leukemias] <- "Leukemia" 
merged_df$Dxgroup[merged_df$Disease_type %in% lymphomas] <- "Lymphoma" 
merged_df$Dxgroup[merged_df$Disease_type %in% solids] <- "Solid" 
```

```r
table(merged_df$Dxgroup)
```

```r
ggplot(data = merged_df[merged_df$Dxgroup == "Leukemia",], 
       aes( y= CPKR, x = cycle)) + geom_boxplot() + mytheme + geom_jitter()
```

```r
colnames(merged_df)
```

```r
ggplot(data = merged_df, aes( y= `Naïve%`, x = log10(observed_Shannon))) + geom_point() + mytheme + geom_jitter()
```

```r
baselines <- merged_df[merged_df$cycle == "01",]
```

```r
baselinesnaomit <- baselines[!is.na(baselines$Foldgroup),]
```

```r
wilcox.test(baselinesnaomit$`Naïve%` ~ baselinesnaomit$Foldgroup)
```

```r
wilcox.test(baselinesnaomit$observed_Shannon ~ baselinesnaomit$Foldgroup)
```
